```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per_capita_income')
plt.scatter(df['year'], df['per_capita_income'], color='red', marker='+')
```

```python
new_df = df.drop('per_capita_income',axis='columns')
new_df
```

```python
per_capita_income = df.per_capita_income
per_capita_income
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,per_capita_income)
```

```python
reg.predict([[2001]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
2001*828.46507522 + -1632210.7578554575
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per_capita_income')
plt.scatter(df['year'], df['per_capita_income'], color='red', marker='+')
plt.plot(df['year'], reg.predict(df[['year']]), color = 'blue')
```

```python
year_df = pd.read_csv("year.csv")
year_df.head(3)
```

```python
p = reg.predict(year_df)
p
```
